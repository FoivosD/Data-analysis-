```python
# removing the null values
df=df.dropna()
```

```python
#Transform date column from object to date time
df['date']=pd.to_datetime(df['date'])
df.head(2)
```

```python
#The date column has changed to a datetime type
df.info()
```

```python
#extract year from date column
df['year']=df['date'].dt.year
```

```python
df.head()
```

```python
#Extract month from date column
df['month']=df['date'].dt.month
df.head()
```

```python
#Display records with 0 crimes
df[df['no_of_crimes']==0]
```

```python
#Count the years in the data
df['year'].value_counts()
```

```python
#How many crimes we have in city of London
df[(df['area']=="city of london")& (df['no_of_crimes']) !=0]
```

```python
#Average price per Area across all years
df.groupby('area')['average_price'].mean().round(0)
```

```python
#Let's find more about the prices per year starting with the minimum price
df.groupby('year')['average_price'].min().round(0)
```

```python
#Max prices per year
df.groupby('year')['average_price'].max().round(0)
```

```python
#And finally the mean
df.groupby('year')["average_price"].mean().round(0)
```

```python
#Minimum number of crimes per area
df.groupby('area')['no_of_crimes'].min()
```
